```python
# Synthetic Employee Performance Analysis
# Verification email: 22f3002903@ds.study.iitm.ac.in

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ---------------- 1) Generate synthetic dataset ----------------
np.random.seed(42)
n = 100

employee_ids = [f"EMP{str(i).zfill(3)}" for i in range(1, n + 1)]
departments = ["HR", "Sales", "Marketing", "Engineering", "Finance", "Operations"]
regions = ["North", "South", "East", "West"]

dept_probs = [0.12, 0.22, 0.14, 0.26, 0.12, 0.14]  # weights for distribution
region_probs = [0.28, 0.24, 0.24, 0.24]

df = pd.DataFrame({
    "EmployeeID": employee_ids,
    "Department": np.random.choice(departments, size=n, p=dept_probs),
    "Region": np.random.choice(regions, size=n, p=region_probs),
    "PerformanceScore": np.clip(np.random.normal(3.5, 0.8, size=n), 1, 5).round(2),
    "YearsOfService": np.clip(np.random.poisson(6, size=n), 0, 35)
})

# ---------------- 2) Frequency count for HR ----------------
hr_count = (df["Department"] == "HR").sum()
print(f"Number of employees in HR: {hr_count}")

# ---------------- 3) Histogram (distribution across Departments) ----------------
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Department", order=departments, palette="Set2")
plt.title("Distribution of Employees by Department")
plt.xlabel("Department")
plt.ylabel("Number of Employees")

# ---------------- 4) Show the plot ----------------
plt.show()
```
